```python
import pandas as pd
```

```python
df = pd.read_csv("/content/vehical.csv")
df
```

```python
dummies = pd.get_dummies(df.Car_Model)
```

```python
merged = pd.concat([df,dummies],axis ='columns')
merged
```

```python
final = merged.drop(['Car_Model','Mercedez Benz C class'],axis='columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
X = final.drop(['Sell_Price'],axis='columns')
X
```

```python
y =final.Sell_Price
y
```

```python
model.fit(X,y)
```

```python
model.predict([[69000,6,0,1]])
```

```python
model.score(X,y)
```
