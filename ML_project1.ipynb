```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python

sonar_data = pd.read_csv('/content/sonarData.csv',header=None); sonar_data.head()
```

```python
sonar_data.shape
```

```python
sonar_data[60].value_counts()
```

```python
#seperating data and labels
X=sonar_data.drop(columns=60,axis=1)
Y =sonar_data[60]
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.1,stratify=Y,random_state=1)
```

```python
model=LogisticRegression()
```

```python
y_train.shape
```

```python
model.fit(x_train,y_train)
```

```python
x_train_prediction=model.predict(x_test)
data_accuracy = accuracy_score(x_train_prediction,y_test)
```

```python
print("accuracy of the data model : ", data_accuracy)
```

```python
x_train_prediction_train=model.predict(x_train)
data_accuracy_train = accuracy_score(x_train_prediction_train,y_train)
print("accuracy of the data model : ", data_accuracy_train)
```
